```python
import json
import numpy as np
import cv2
import os
import imutils
```

```python
folder_path = 'D:/Intozi Task 4/IMG/'
jsonpath='D:/Intozi Task 4/ANNO/'
```

```python
for file in os.listdir(folder_path):
    with open(f'{jsonpath}/'+str(file[0:-3:])+'json') as f:
        data = json.load(f)
    points = data['shapes'][0]['points']
    pts = np.array(points,np.int32)
    pts = pts.reshape((-1, 1, 2))
    path = f'{folder_path}/'+str(file)
    image = cv2.imread(path)
    img = np.zeros(image.shape, np.uint8) 
    cv2.fillPoly(img, [pts], (255, 255, 255))
    cv2.imwrite('finalimg/'+str(file[0:-4:])+'mask'+'.jpg',img)
    print ("done",file)
        
```
